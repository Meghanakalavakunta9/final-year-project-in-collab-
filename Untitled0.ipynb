```python
# import boto3
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# # Step 1: Load Dataset from S3
# s3 = boto3.client('s3')
# bucket_name = '<your-s3-bucket-name>'  # Replace with actual bucket name
# file_key = '<heart_cleveland.csv>'    # Replace with actual dataset filename in S3

# # Download dataset
# s3.download_file(bucket_name, file_key, 'heart_cleveland.csv')

# Load dataset
heart_data = pd.read_csv('/content/heart_cleveland.csv')

# Step 2: Data Preprocessing
x = heart_data.drop(columns='target', axis=1)
y = heart_data['target']

# Splitting the dataset into training and testing sets (80-20 split)
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, stratify=y, random_state=42
)

# Step 3: XGBoost Model with Regularization to Reduce Overfitting
model = XGBClassifier(
    n_estimators=100,       # Number of trees
    max_depth=3,            # Limits complexity of trees
    learning_rate=0.1,      # Controls how much to adjust each tree
    subsample=0.8,          # Uses 80% of the data per tree
    colsample_bytree=0.8,   # Uses 80% of features per tree
    reg_lambda=1,           # L2 regularization (reduces overfitting)
    reg_alpha=0.5,          # L1 regularization (encourages sparsity)
    eval_metric="logloss",
    use_label_encoder=False
)

# Step 4: Model Training with Early Stopping
model.fit(
    x_train, y_train,
    eval_set=[(x_test, y_test)],
    early_stopping_rounds=10,  # Stops training if no improvement in 10 rounds
    verbose=False
)

# Step 5: Model Evaluation
x_train_pred = model.predict(x_train)
training_accuracy = accuracy_score(y_train, x_train_pred)
print("Training Accuracy (Reduced Overfitting):", training_accuracy)

x_test_pred = model.predict(x_test)
testing_accuracy = accuracy_score(y_test, x_test_pred)
print("Testing Accuracy:", testing_accuracy)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# Loading the csv data to a Pandas Dataframe
heart_data = pd.read_csv('/content/heart_cleveland.csv')

# print first 5 rows of the dataset
heart_data.head()

# print last 5 rows of the dataset
heart_data.tail()

# The number of rows and columns in the dataset
heart_data.shape

# Getting some info about the data
heart_data.info()

# Checking for missing values
heart_data.isnull().sum()

# statistical measures about the data
heart_data.describe()

# Checking the distribution of target variable (target values)
# (1) ---------> Defective heart
# (0) ---------> Healthy heart
heart_data['target'].value_counts()

# Splitting the features and target values
x = heart_data.drop(columns='target',axis=1)
y = heart_data['target']
print(x)

print(y)

# splitting the data into training data and testing data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify = y,random_state = 2)
print(x.shape,x_train.shape,x_test.shape)

# Model initialization
xgb_model = XGBClassifier()

# Training the model
xgb_model.fit(x_train, y_train)

# Accuracy on training data
x_train_pred_xgb = xgb_model.predict(x_train)
train_acc_xgb = accuracy_score(y_train, x_train_pred_xgb)
print("XGBoost - Accuracy on Training Data: ", train_acc_xgb)

# Accuracy on testing data
x_test_pred_xgb = xgb_model.predict(x_test)
test_acc_xgb = accuracy_score(y_test, x_test_pred_xgb)
print("XGBoost - Accuracy on Testing Data: ", test_acc_xgb)
```
